# Loan Project Example:

A college student is considering borrowing $50,000 to cover expenses
for their upcoming senior year.

The loan is to be paid back in 5 yearly installments of $12,000 over the
first 5 years of the student's professional career

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Constants
loan_amount = 50000
annual_payment = 12000
years = 6
```

```python
# Create a list to store the data
data = {'Year': [], 'Transaction': []}

# Populate the data list
for year in range(years + 1):
    if year == 0:
        transaction = loan_amount
    elif year < 2:
        transaction = 0
    else:
        transaction = -annual_payment
    data['Year'].append(year)
    data['Transaction'].append(transaction)

# Create a DataFrame
Loan_df = pd.DataFrame(data)

# Print the table
Loan_df
```

```python
# Create a figure and axis for the cash flow diagram
fig, ax = plt.subplots(figsize=(15, 7))

# Create arrows for each year's cash flow
for year in range(years + 1):
    if year == 0:
        cash_flow = loan_amount
    elif year < 2:
        cash_flow = 0
    else:
        cash_flow = -annual_payment
    ax.arrow(year, 0, 0, cash_flow, head_width=0.2, head_length=1000, fc='green', ec='green')
    ax.text(year, cash_flow / 2, f"${cash_flow}", ha='center', va='center', color='green')

# Set labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Cash Flow')
ax.set_title('Cash Flow Diagram')

# Add a horizontal line at y=0 for reference
ax.axhline(0, color='red', linestyle='--')

# Set the x-axis limits to include all years
ax.set_xlim(-1, years + 1)

# Show the cash flow diagram
plt.grid(True)
plt.show()
```
