#  Calculate Time to Double Principal with Simple and Compound Interest

This project calculates and visualizes the growth of a principal amount over a specified number of years using both simple and compound interest. It provides a clear comparison between the two interest methods and helps users understand how interest affects their investments over time.

Principal (p): $25,000

Interest Rate(r): 10%

Years: 10

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Constants
P = 25000  # Principal
r = 0.10   # Annual interest rate (10%)
years = 10  # Number of years to track
```

```python
# Initialize lists to store data
years_list = list(range(years + 1))
principal_simple = []
principal_compound = []
```

```python
# Calculate principal increase for each year
for year in years_list:
    principal_simple.append(P + (P * r * year))  # Simple Interest formula
    principal_compound.append(P * (1 + r) ** year)  # Compound Interest formula
```

```python
# Create a DataFrame to store the data
data = {'Year': years_list, 'Simple Interest': principal_simple, 'Compound Interest': principal_compound}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Format the columns with 2 decimal places and add a dollar sign
df[['Simple Interest', 'Compound Interest']] = df[['Simple Interest', 'Compound Interest']].applymap('${:.2f}'.format)
```

```python
df
```

```python
# Plot the principal increase over the years
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Principal (Simple Interest)'], label='Simple Interest', marker='o')
plt.plot(df['Year'], df['Principal (Compound Interest)'], label='Compound Interest', marker='o')
plt.xlabel('Year')
plt.ylabel('Principal Amount')
plt.title('Principal Increase Over the Years (Simple vs. Compound Interest)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
